#### Problem statement

> Predicting the quality of dataset on the basis of features.
Dataset: https://www.kaggle.com/datasets/rajyellow46/wine-quality/download?datasetVersionNumber=1

Feature Description

        * volatile acidity :   Volatile acidity is the gaseous acids present in wine.

        * fixed acidity :   Primary fixed acids found in wine are tartaric, succinic, citric, and malic

        * residual sugar :   Amount of sugar left after fermentation.

        * citric acid :    It is weak organic acid, found in citrus fruits naturally.

        * chlorides :   Amount of salt present in wine.

        * free sulfur dioxide :   So2 is used for prevention of wine by oxidation and microbial spoilage.

        * total sulfur dioxide 

        * pH :   In wine pH is used for checking acidity

        * density 

        * sulphates :    Added sulfites preserve freshness and protect wine from oxidation, and bacteria.
        * alcohol :   Percent of alcohol present in wine.

#### Removing missing value with mean of that feature

```python
for col in data.columns:
    if data[col].isna().sum()>0:
        data[col]=data[col].fillna(data[col].mean())
```

```python
data=data.drop(columns="free sulfur dioxide",axis=1)
```

```python
data.groupby(['type']).count()
```

```python
data['type']=data['type'].replace({'red':0,'white':1})
```

```python
data.head()
```

```python
data.tail()
```

```python
y=data['quality']
x=data.drop(columns='quality',axis=0)
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=70,test_size=0.2)
```

```python
sc=StandardScaler()
sc.fit(x_train)
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

### DecisionTreeClassifier Model

```python
dt_model=DecisionTreeClassifier(criterion='entropy',random_state=70)
dt_model.fit(x_train,y_train)
print("Accuracy on Train data:",dt_model.score(x_train,y_train)*100)
print("Accuracy on Test data:",dt_model.score(x_test,y_test)*100)
```

### RandomForestClassifier Model

```python
rd_model=RandomForestClassifier(criterion='gini',random_state=70)
rd_model.fit(x_train,y_train)
print("Accuracy on Train data:",rd_model.score(x_train,y_train)*100)
print("Accuracy on Test data:",rd_model.score(x_test,y_test)*100)
```

### ExtraTreeClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
et_model = ExtraTreesClassifier()
et_model.fit(x_train,y_train)
et_model.score(x_test,y_test)
```

### SVC

```python
from sklearn.svm import SVC
svc= SVC(kernel='rbf',gamma='auto',C=1)
svc.fit(x_train,y_train)
svc.score(x_test,y_test)
```
